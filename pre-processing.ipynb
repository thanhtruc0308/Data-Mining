```python
import pandas as pd
import numpy as np

full_df = pd.read_csv("./dataset/data.csv", encoding = "unicode_escape")
full_df.head()
```

```python
missing_data = pd.DataFrame(full_df.isnull().sum(), columns=["Count"])
missing_data.sort_values(by=["Count"])
```

```python
full_df = full_df.dropna(subset=["CustomerID"])
full_df["CustomerID"]
```

```python
full_df["CustomerID"] = full_df["CustomerID"].apply(lambda x: str(x)[:-2])
full_df["CustomerID"]
```

```python
full_df[full_df["UnitPrice"]==0]
```

```python
full_df.drop(full_df[full_df["UnitPrice"]==0].index, inplace=True)
```

```python
all_stock = list(set(full_df['StockCode']))
all_stock.sort()
```

```python
basket = full_df.groupby(['InvoiceNo','CustomerID'], as_index= False).agg({'StockCode': lambda s: list(set(s))}) # grouping product from the same invoice. 
basket
```

```python
for item in all_stock:
    basket[item] = item
basket
```

```python
def binarySearch(arr, l, r, x):
    x = str(x)
    if (r >= l):         
        mid = int(l + (r - l) / 2)

        if (str(arr[mid]) == x): 
            arr[mid] = 'True'
        if (mid > l and str(arr[mid - 1]) == x):
            arr[mid-1] = 'True'
        if (mid < r and str(arr[mid + 1]) == x):
            arr[mid+1] = 'True'

        if (str(arr[mid]) > x):
            return binarySearch(arr, l, mid - 2, x)
         
        return binarySearch(arr, mid + 2, r, x)

    return arr
```

```python
data = np.array(basket)
```

```python
for row in range(len(data)):
    for item in data[row][2]:
        binarySearch(data[row],3,len(all_stock) - 1,item)
        
```

```python
for row in range(len(data)):
    for col in range(3,len(all_stock)):
        if data[row][col] != 'True':
            data[row][col] = 'False'
    
```

```python
data = pd.DataFrame(data,columns=basket.columns)
data = data.drop(columns=['StockCode','BANK CHARGES','C2','CRUK','D','DOT','M','PADS','POST'],axis=1)
data
```

```python
data.to_csv('Preprocess.csv', index=False)
```
